```python
import os, shutil

# find all .hvo files in the directory

dir_ = "data/triple_streams/candombe_hvo"
hvo_files = [f for f in os.listdir(dir_) if f.endswith('.hvo')]
print(hvo_files)
```

```python
loaded_hvo_seqs_candombe = []

for hvo_file in hvo_files:
    hvo_seq = load_HVO_Sequence_from_file(os.path.join(dir_, hvo_file))
    loaded_hvo_seqs_candombe.append(hvo_seq)
    print(f"Loaded {hvo_file} events with {hvo_seq.hvo.shape[0]} steps, i.e. {hvo_seq.hvo.shape[0] // 32} 2bar phrases.")
```
